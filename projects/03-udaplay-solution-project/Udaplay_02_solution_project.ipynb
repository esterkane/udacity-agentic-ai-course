Deliverables covered here:
- `retrieve_game` tool for local vector search
- `evaluate_retrieval` tool for deciding whether RAG is enough
- `game_web_search` tool using Tavily fallback
- a stateful agent class with a clear state-machine workflow
- persistent long-term memory for facts learned from web searches
- example runs with tool usage, reasoning summary, citations, confidence, and structured output

## Data models

```python
class RetrievedGame(BaseModel):
    id: str
    name: str
    platform: str
    genre: str
    publisher: str
    year_of_release: int
    description: str
    source: str
    distance: float
    relevance_score: float


class EvaluationReport(BaseModel):
    useful: bool = Field(description="Whether the retrieved documents are enough to answer the user question")
    confidence: float = Field(ge=0.0, le=1.0, description="Confidence that local RAG can answer the question")
    explanation: str
    recommended_action: str = Field(description="answer_from_internal_knowledge or search_web")


class SourceCitation(BaseModel):
    label: str
    url: Optional[str] = None
    source_type: str
    snippet: Optional[str] = None


class AgentAnswer(BaseModel):
    question: str
    final_answer: str
    confidence: float
    answer_source: str
    citations: List[SourceCitation]
    structured_data: Dict[str, Any] = Field(default_factory=dict)
    tool_usage: List[Dict[str, Any]] = Field(default_factory=list)
    reasoning_summary: List[str] = Field(default_factory=list)
```

## Utility helpers

```python
def _safe_json_loads(value: str, default):
    try:
        return json.loads(value)
    except Exception:
        return default


def _normalise(text: str) -> str:
    # Keep unicode letters such as "Ã©" in PokÃ©mon while removing punctuation.
    return re.sub(r"[_\W]+", " ", text.lower(), flags=re.UNICODE).strip()


def _important_tokens(text: str) -> set[str]:
    stopwords = {
        "the", "a", "an", "of", "and", "or", "for", "to", "was", "were", "is", "on",
        "when", "what", "which", "who", "released", "release", "platform", "platforms",
        "game", "games", "video", "developed", "developer", "publisher", "published",
        "did", "in", "with", "from", "listed", "one", "it"
    }
    return {token for token in _normalise(text).split() if token and token not in stopwords}


# These tokens often describe where/how a game runs rather than the game identity.
# Filtering them avoids accepting a wrong game simply because both mention "PlayStation 5".
PLATFORM_OR_CONTEXT_TOKENS = {
    "playstation", "ps5", "ps4", "ps3", "ps2", "xbox", "series", "switch", "nintendo",
    "gamecube", "wii", "nes", "snes", "pc", "windows", "steam", "console", "consoles",
    "boy", "color", "advance", "released", "release", "platform", "platforms", "right",
    "now", "currently", "latest", "upcoming", "announced", "working"
}


def _question_identity_tokens(question: str) -> set[str]:
    """Return tokens that likely identify the requested game/topic, not only the platform."""
    tokens = _important_tokens(question) - PLATFORM_OR_CONTEXT_TOKENS
    # Keep one-letter roman-title tokens such as "x" in "Mortal Kombat X".
    return {token for token in tokens if token}


def _title_overlap(question: str, title: str) -> float:
    title_tokens = _important_tokens(title)
    if not title_tokens:
        return 0.0
    question_tokens = _important_tokens(question)
    return len(title_tokens & question_tokens) / len(title_tokens)


def _local_support_score(question: str, doc: Dict[str, Any]) -> float:
    """How strongly the retrieved local document supports the specific user question."""
    question_tokens = _question_identity_tokens(question)
    if not question_tokens:
        return 0.0

    searchable_doc_text = " ".join([
        str(doc.get("name", "")),
        str(doc.get("genre", "")),
        str(doc.get("publisher", "")),
        str(doc.get("description", "")),
    ])
    doc_tokens = _important_tokens(searchable_doc_text)
    shared = question_tokens & doc_tokens
    return round(len(shared) / len(question_tokens), 4)


def _rank_retrieved_docs(question: str, docs: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
    """Rank Chroma results with an extra lexical/entity-support check."""
    ranked = []
    for doc in docs:
        enriched = dict(doc)
        title_overlap = round(_title_overlap(question, doc.get("name", "")), 4)
        support_score = _local_support_score(question, doc)
        relevance_score = float(doc.get("relevance_score", 0.0))

        # Entity/content support matters more than raw vector distance for factual QA.
        combined_score = round(max(title_overlap, support_score) + (0.15 * relevance_score), 4)

        enriched["title_overlap"] = title_overlap
        enriched["local_support_score"] = support_score
        enriched["combined_support_score"] = combined_score
        ranked.append(enriched)

    return sorted(
        ranked,
        key=lambda item: (
            item.get("combined_support_score", 0.0),
            item.get("local_support_score", 0.0),
            item.get("title_overlap", 0.0),
            item.get("relevance_score", 0.0),
        ),
        reverse=True,
    )


def _distance_to_relevance(distance: float) -> float:
    # Chroma cosine distance is lower-is-better. This conversion gives a simple 0..1 confidence-like score.
    return round(max(0.0, min(1.0, 1.0 / (1.0 + float(distance)))), 4)


def load_long_term_memory() -> List[Dict[str, Any]]:
    if not MEMORY_PATH.exists():
        return []
    return _safe_json_loads(MEMORY_PATH.read_text(encoding="utf-8"), [])


def save_long_term_memory(records: List[Dict[str, Any]]) -> None:
    MEMORY_PATH.write_text(json.dumps(records, indent=2, ensure_ascii=False), encoding="utf-8")


def remember_web_result(question: str, web_result: Dict[str, Any]) -> None:
    records = load_long_term_memory()
    records.append({
        "timestamp": int(time.time()),
        "question": question,
        "answer": web_result.get("answer"),
        "sources": web_result.get("sources", []),
    })
    save_long_term_memory(records)


def search_long_term_memory(question: str, limit: int = 3) -> List[Dict[str, Any]]:
    # Lightweight persistent memory search. It is intentionally simple and transparent for this project.
    records = load_long_term_memory()
    query_tokens = _question_identity_tokens(question) or _important_tokens(question)
    scored = []
    for record in records:
        haystack = f"{record.get('question', '')} {record.get('answer', '')}"
        overlap = len(query_tokens & _important_tokens(haystack))
        if overlap > 0:
            scored.append((overlap, record))
    scored.sort(key=lambda item: item[0], reverse=True)
    return [record for _, record in scored[:limit]]
```

## Tool 1 â€” retrieve_game

```python
@tool
def retrieve_game(query: str, n_results: int = 4) -> str:
    """
    Search the local UdaPlay vector database for video game information.

    Args:
        query: Natural language question about a video game, publisher, platform, genre, or release year.
        n_results: Maximum number of local game records to return.

    Returns:
        JSON list of matching game records. Each record includes title, platform, genre, publisher,
        release year, description, local source file, vector distance, relevance score, and support scores.
    """
    results = collection.query(
        query_texts=[query],
        n_results=n_results,
        include=["documents", "metadatas", "distances"],
    )

    rows: List[Dict[str, Any]] = []
    for doc_id, metadata, distance in zip(
        results["ids"][0],
        results["metadatas"][0],
        results["distances"][0],
    ):
        rows.append({
            "id": doc_id,
            "name": metadata["Name"],
            "platform": metadata["Platform"],
            "genre": metadata["Genre"],
            "publisher": metadata["Publisher"],
            "year_of_release": int(metadata["YearOfRelease"]),
            "description": metadata["Description"],
            "source": metadata.get("source_file", f"games/{doc_id}.json"),
            "distance": round(float(distance), 4),
            "relevance_score": _distance_to_relevance(float(distance)),
        })

    rows = _rank_retrieved_docs(query, rows)
    return json.dumps(rows, indent=2, ensure_ascii=False)
```

## Tool 2 â€” evaluate_retrieval

```python
@tool
def evaluate_retrieval(question: str, retrieved_docs: str) -> str:
    """
    Evaluate whether local vector search results are enough to answer the user's question.

    Args:
        question: Original user question.
        retrieved_docs: JSON string returned by retrieve_game.

    Returns:
        JSON EvaluationReport with useful, confidence, explanation, and recommended_action.
    """
    docs = _safe_json_loads(retrieved_docs, [])
    if not docs:
        report = EvaluationReport(
            useful=False,
            confidence=0.0,
            explanation="No local documents were retrieved.",
            recommended_action="search_web",
        )
        return report.model_dump_json(indent=2)

    question_lower = question.lower()
    current_info_keywords = ["right now", "currently", "latest", "working on", "upcoming", "announced"]
    asks_for_current_info = any(keyword in question_lower for keyword in current_info_keywords)

    top_doc = docs[0]
    top_score = float(top_doc.get("relevance_score", 0.0))
    title_overlap = float(top_doc.get("title_overlap", _title_overlap(question, top_doc.get("name", ""))))
    support_score = float(top_doc.get("local_support_score", _local_support_score(question, top_doc)))
    identity_tokens = _question_identity_tokens(question)

    # The local dataset is static. Current/upcoming work should normally go to web search.
    if asks_for_current_info:
        report = EvaluationReport(
            useful=False,
            confidence=min(top_score, 0.45),
            explanation=(
                "The question asks for current or upcoming information. The local game JSON files are static "
                "and should not be treated as current market/news data."
            ),
            recommended_action="search_web",
        )
        return report.model_dump_json(indent=2)

    # A pure vector score can be misleading. Example: a question about Mortal Kombat X + PS5 can retrieve
    # a PlayStation 5 game even when the title is unrelated. Require actual title/content support.
    has_specific_identity = len(identity_tokens) >= 2
    useful = (
        title_overlap >= 0.5
        or support_score >= 0.55
        or (top_score >= 0.78 and support_score >= 0.25)
    )

    if has_specific_identity and support_score < 0.35 and title_overlap < 0.5:
        useful = False

    confidence = round(max(title_overlap, support_score, min(top_score, 0.75 if useful else 0.45)), 4)

    if useful:
        explanation = (
            f"The best local result is '{top_doc.get('name')}' with relevance score {top_score}, "
            f"title overlap {round(title_overlap, 2)}, and local support score {round(support_score, 2)}. "
            "This is enough for a factual answer from the local dataset."
        )
        action = "answer_from_internal_knowledge"
    else:
        explanation = (
            f"The local retrieval result is not specific enough. Best candidate: '{top_doc.get('name')}', "
            f"relevance score {top_score}, title overlap {round(title_overlap, 2)}, "
            f"local support score {round(support_score, 2)}. A web search is safer."
        )
        action = "search_web"

    report = EvaluationReport(
        useful=useful,
        confidence=confidence,
        explanation=explanation,
        recommended_action=action,
    )
    return report.model_dump_json(indent=2)
```

## Tool 3 â€” game_web_search

```python
@tool
def game_web_search(question: str, max_results: int = 3) -> str:
    """
    Search the web for current or missing video game information using Tavily.

    Args:
        question: Natural language question about the video game industry.
        max_results: Maximum number of web pages to retrieve.

    Returns:
        JSON object with a web answer and citation sources.
    """
    search_query = f"{question} video game official source release platform publisher developer"
    response = tavily_client.search(
        query=search_query,
        search_depth="advanced",
        max_results=max_results,
        include_answer=True,
        include_raw_content=False,
    )

    sources = []
    for result in response.get("results", []):
        sources.append({
            "title": result.get("title"),
            "url": result.get("url"),
            "content": (result.get("content") or "")[:900],
            "score": result.get("score"),
        })

    payload = {
        "answer": response.get("answer"),
        "sources": sources,
    }
    return json.dumps(payload, indent=2, ensure_ascii=False)
```

## Agent implementation

```python
class UdaPlayState(str, Enum):
    START = "start"
    CHECK_MEMORY = "check_memory"
    RETRIEVE = "retrieve_game"
    EVALUATE = "evaluate_retrieval"
    WEB_SEARCH = "game_web_search"
    PERSIST_MEMORY = "persist_memory"
    REPORT = "report"
    DONE = "done"


@dataclass
class UdaPlaySessionState:
    """Short-term state for one independent conversation session."""

    session_id: str
    history: List[Dict[str, Any]] = field(default_factory=list)
    last_question: Optional[str] = None
    last_answer: Optional[Dict[str, Any]] = None
    created_at: int = field(default_factory=lambda: int(time.time()))
    updated_at: int = field(default_factory=lambda: int(time.time()))


class UdaPlayResearchAgent:
    """Stateful research agent for a single video-game QA session.

    Each instance owns one short-term memory object. The short-term memory is
    used to resolve follow-up questions like "What platform was it released
    for?" by looking at the previous answer in the same session only.

    Workflow:
    START -> CHECK_MEMORY -> RETRIEVE -> EVALUATE ->
    either REPORT from internal data, or WEB_SEARCH -> PERSIST_MEMORY -> REPORT.
    """

    def __init__(self, session_id: str = "default"):
        self.state = UdaPlaySessionState(session_id=session_id)

    def ask(self, question: str) -> AgentAnswer:
        original_question = question.strip()
        retrieval_question, context_info = self._contextualize_question(original_question)

        tool_usage: List[Dict[str, Any]] = []
        reasoning_summary: List[str] = []

        current_state = UdaPlayState.START
        reasoning_summary.append(f"Started a UdaPlay research workflow for session '{self.state.session_id}'.")
        tool_usage.append({
            "state": current_state.value,
            "tool": "short_term_memory_context",
            "session_id": self.state.session_id,
            "used_short_term_memory": context_info["used_short_term_memory"],
            "previous_subject": context_info.get("previous_subject"),
            "retrieval_question": retrieval_question,
        })

        if context_info["used_short_term_memory"]:
            reasoning_summary.append(
                "Resolved the follow-up question with this session's short-term memory: "
                f"'{context_info['previous_subject']}'."
            )
        else:
            reasoning_summary.append("The question did not need previous session context.")

        current_state = UdaPlayState.CHECK_MEMORY
        memory_hits = search_long_term_memory(retrieval_question)
        tool_usage.append({
            "state": current_state.value,
            "tool": "search_long_term_memory",
            "session_id": self.state.session_id,
            "result_count": len(memory_hits),
        })
        if memory_hits:
            reasoning_summary.append("Found related long-term memory from earlier web searches, but still checked local RAG first.")
        else:
            reasoning_summary.append("No relevant long-term memory found; checking local game dataset first.")

        current_state = UdaPlayState.RETRIEVE
        retrieved_docs_json = retrieve_game(retrieval_question)
        retrieved_docs = _safe_json_loads(retrieved_docs_json, [])
        tool_usage.append({
            "state": current_state.value,
            "tool": "retrieve_game",
            "session_id": self.state.session_id,
            "result_count": len(retrieved_docs),
        })

        current_state = UdaPlayState.EVALUATE
        evaluation_json = evaluate_retrieval(retrieval_question, retrieved_docs_json)
        evaluation = EvaluationReport.model_validate_json(evaluation_json)
        tool_usage.append({
            "state": current_state.value,
            "tool": "evaluate_retrieval",
            "session_id": self.state.session_id,
            "useful": evaluation.useful,
            "confidence": evaluation.confidence,
            "recommended_action": evaluation.recommended_action,
        })
        reasoning_summary.append(evaluation.explanation)

        if evaluation.useful:
            current_state = UdaPlayState.REPORT
            answer = self._build_internal_answer(
                original_question,
                retrieval_question,
                retrieved_docs,
                evaluation,
                tool_usage,
                reasoning_summary,
            )
        else:
            current_state = UdaPlayState.WEB_SEARCH
            web_json = game_web_search(retrieval_question)
            web_result = _safe_json_loads(web_json, {"answer": None, "sources": []})
            tool_usage.append({
                "state": current_state.value,
                "tool": "game_web_search",
                "session_id": self.state.session_id,
                "result_count": len(web_result.get("sources", [])),
            })
            reasoning_summary.append("Local knowledge was not enough, so the workflow used Tavily web search as fallback.")

            current_state = UdaPlayState.PERSIST_MEMORY
            remember_web_result(original_question, web_result)
            tool_usage.append({
                "state": current_state.value,
                "tool": "remember_web_result",
                "session_id": self.state.session_id,
                "memory_file": str(MEMORY_PATH),
            })
            reasoning_summary.append("Stored the web-search result in persistent long-term memory for future related queries.")

            current_state = UdaPlayState.REPORT
            answer = self._build_web_answer(
                original_question,
                retrieval_question,
                web_result,
                evaluation,
                tool_usage,
                reasoning_summary,
            )

        self.state.last_question = original_question
        self.state.last_answer = answer.model_dump()
        self.state.updated_at = int(time.time())
        self.state.history.append({
            "session_id": self.state.session_id,
            "question": original_question,
            "retrieval_question": retrieval_question,
            "answer": answer.model_dump(),
            "timestamp": self.state.updated_at,
        })

        return answer

    def _contextualize_question(self, question: str) -> tuple[str, Dict[str, Any]]:
        """Use this session's short-term memory to resolve follow-up questions.

        This is intentionally lightweight and transparent. It does not share
        state across sessions. A follow-up in `pokemon-session` can only use the
        previous answer from `pokemon-session`, not from another session.
        """

        previous_subject = self._previous_subject()
        lower_question = f" {question.lower()} "
        follow_up_markers = [
            " it ", " its ", " that game ", " this game ", " the game ",
            " that one ", " released for", " released on", " platform was",
            " platform is", " publisher", " published it", " developed it",
        ]
        looks_like_follow_up = any(marker in lower_question for marker in follow_up_markers)

        if previous_subject and looks_like_follow_up and previous_subject.lower() not in lower_question:
            contextual_question = (
                f"{question}\n\n"
                f"Session context: In session '{self.state.session_id}', the previous answer was about "
                f"'{previous_subject}'. Interpret pronouns such as 'it', 'that game', or 'the game' as "
                f"'{previous_subject}'."
            )
            return contextual_question, {
                "used_short_term_memory": True,
                "previous_subject": previous_subject,
            }

        return question, {
            "used_short_term_memory": False,
            "previous_subject": previous_subject,
        }

    def _previous_subject(self) -> Optional[str]:
        if not self.state.last_answer:
            return None

        structured = self.state.last_answer.get("structured_data", {})
        matched_game = structured.get("matched_game")
        if isinstance(matched_game, dict) and matched_game.get("name"):
            return matched_game["name"]

        # For web answers, keep the original question as loose context. This is
        # safer than trying to extract entities from arbitrary web text.
        return self.state.last_question

    def _build_internal_answer(
        self,
        question: str,
        retrieval_question: str,
        docs: List[Dict[str, Any]],
        evaluation: EvaluationReport,
        tool_usage: List[Dict[str, Any]],
        reasoning_summary: List[str],
    ) -> AgentAnswer:
        top = docs[0]
        final_answer = (
            f"Based on the local UdaPlay game dataset, {top['name']} was released for "
            f"{top['platform']} in {top['year_of_release']}. It is a {top['genre']} game "
            f"published by {top['publisher']}. Description: {top['description']}"
        )

        citations = [
            SourceCitation(
                label=f"Local dataset: {top['source']}",
                source_type="local_vector_db",
                snippet=f"{top['name']} | {top['platform']} | {top['publisher']} | {top['year_of_release']}",
            )
        ]

        structured_data = {
            "session_id": self.state.session_id,
            "original_question": question,
            "retrieval_question": retrieval_question,
            "matched_game": top,
            "retrieval_evaluation": evaluation.model_dump(),
            "alternative_local_matches": docs[1:],
        }

        return AgentAnswer(
            question=question,
            final_answer=final_answer,
            confidence=evaluation.confidence,
            answer_source="internal_rag",
            citations=citations,
            structured_data=structured_data,
            tool_usage=tool_usage,
            reasoning_summary=reasoning_summary,
        )

    def _build_web_answer(
        self,
        question: str,
        retrieval_question: str,
        web_result: Dict[str, Any],
        evaluation: EvaluationReport,
        tool_usage: List[Dict[str, Any]],
        reasoning_summary: List[str],
    ) -> AgentAnswer:
        sources = web_result.get("sources", [])
        citations = [
            SourceCitation(
                label=source.get("title") or source.get("url") or "Web source",
                url=source.get("url"),
                source_type="web",
                snippet=source.get("content"),
            )
            for source in sources
        ]

        web_answer = web_result.get("answer")
        if not web_answer and sources:
            web_answer = sources[0].get("content")
        if not web_answer:
            web_answer = "I could not find a reliable web answer for this query."

        confidence = 0.72 if sources else 0.35
        if len(sources) >= 2:
            confidence = 0.82

        return AgentAnswer(
            question=question,
            final_answer=web_answer,
            confidence=confidence,
            answer_source="web_search_fallback",
            citations=citations,
            structured_data={
                "session_id": self.state.session_id,
                "original_question": question,
                "retrieval_question": retrieval_question,
                "web_result": web_result,
                "retrieval_evaluation": evaluation.model_dump(),
            },
            tool_usage=tool_usage,
            reasoning_summary=reasoning_summary,
        )

    def get_history(self) -> List[Dict[str, Any]]:
        return self.state.history


class UdaPlayMultiSessionAgent:
    """Session manager that lets one agent service multiple independent sessions."""

    def __init__(self):
        self.sessions: Dict[str, UdaPlayResearchAgent] = {}

    def get_or_create_session(self, session_id: str) -> UdaPlayResearchAgent:
        if not session_id or not session_id.strip():
            raise ValueError("session_id must be a non-empty string")
        if session_id not in self.sessions:
            self.sessions[session_id] = UdaPlayResearchAgent(session_id=session_id)
        return self.sessions[session_id]

    def ask(self, session_id: str, question: str) -> AgentAnswer:
        session_agent = self.get_or_create_session(session_id)
        return session_agent.ask(question)

    def get_session_history(self, session_id: str) -> List[Dict[str, Any]]:
        return self.get_or_create_session(session_id).get_history()

    def list_sessions(self) -> List[str]:
        return sorted(self.sessions.keys())

    def session_count(self) -> int:
        return len(self.sessions)


def print_agent_report(answer: AgentAnswer) -> None:
    print("=" * 90)
    print(f"QUESTION: {answer.question}")
    print("-" * 90)
    print(f"ANSWER SOURCE: {answer.answer_source}")
    print(f"CONFIDENCE: {answer.confidence}")
    print("\nFINAL ANSWER:")
    print(answer.final_answer)
    print("\nREASONING SUMMARY:")
    for item in answer.reasoning_summary:
        print(f"- {item}")
    print("\nTOOL USAGE:")
    for step in answer.tool_usage:
        print(json.dumps(step, ensure_ascii=False))
    print("\nCITATIONS:")
    for citation in answer.citations:
        if citation.url:
            print(f"- {citation.label}: {citation.url}")
        else:
            print(f"- {citation.label}")
    print("\nSTRUCTURED JSON:")
    print(answer.model_dump_json(indent=2))

```

## Multi-session short-term memory demo

The reviewer asked for explicit session functionality. This cell uses one multi-session agent manager with two independent `session_id` values. Each session receives a follow-up question that relies on the previous answer in that same session.


```python
multi_session_agent = UdaPlayMultiSessionAgent()

# Session A: the follow-up question should resolve "it" to PokÃ©mon Gold and Silver.
pokemon_first = multi_session_agent.ask(
    session_id="pokemon-session",
    question="When was PokÃ©mon Gold and Silver released?",
)
pokemon_followup = multi_session_agent.ask(
    session_id="pokemon-session",
    question="What platform was it released for?",
)

# Session B: same follow-up wording, but the context must stay isolated and resolve to Super Mario 64.
mario_first = multi_session_agent.ask(
    session_id="mario-session",
    question="When was Super Mario 64 released?",
)
mario_followup = multi_session_agent.ask(
    session_id="mario-session",
    question="What platform was it released for?",
)

print("Active sessions:", multi_session_agent.list_sessions())
print("pokemon-session history entries:", len(multi_session_agent.get_session_history("pokemon-session")))
print("mario-session history entries:", len(multi_session_agent.get_session_history("mario-session")))
print("pokemon follow-up used short-term memory:", pokemon_followup.tool_usage[0]["used_short_term_memory"])
print("mario follow-up used short-term memory:", mario_followup.tool_usage[0]["used_short_term_memory"])
print("pokemon follow-up retrieval question:")
print(pokemon_followup.structured_data["retrieval_question"])
print("mario follow-up retrieval question:")
print(mario_followup.structured_data["retrieval_question"])

assert multi_session_agent.session_count() == 2
assert len(multi_session_agent.get_session_history("pokemon-session")) == 2
assert len(multi_session_agent.get_session_history("mario-session")) == 2
assert pokemon_followup.tool_usage[0]["used_short_term_memory"] is True
assert mario_followup.tool_usage[0]["used_short_term_memory"] is True
assert "PokÃ©mon Gold and Silver" in pokemon_followup.structured_data["retrieval_question"]
assert "Super Mario 64" in mario_followup.structured_data["retrieval_question"]

print("Multi-session short-term memory checks passed.")

```
