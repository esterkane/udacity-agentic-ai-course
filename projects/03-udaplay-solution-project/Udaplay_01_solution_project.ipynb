```python
import json
import os
from pathlib import Path
from typing import Dict, List

import chromadb
from chromadb.utils import embedding_functions
from dotenv import load_dotenv
```

```python
# Robust environment loading for Udacity/Vocareum and local runs.
# The project text mentions both `.env` and `config.env`, and Jupyter's working
# directory can differ from the notebook file location. This searches the current
# folder and a few parent folders before asserting.
from dotenv import load_dotenv, find_dotenv
from pathlib import Path


def load_project_env() -> list[str]:
    env_names = [".env", "config.env"]
    roots = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
    loaded_files = []

    for root in roots:
        for env_name in env_names:
            candidate = root / env_name
            if candidate.exists():
                load_dotenv(candidate, override=True)
                loaded_files.append(str(candidate))

    # find_dotenv is useful when the notebook was opened from a nested folder.
    for env_name in env_names:
        discovered = find_dotenv(env_name, usecwd=True)
        if discovered and discovered not in loaded_files:
            load_dotenv(discovered, override=True)
            loaded_files.append(discovered)

    return loaded_files


loaded_env_files = load_project_env()
print("Loaded env files:", loaded_env_files if loaded_env_files else "None found")
print("Current working directory:", Path.cwd())

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# Vocareum/Udacity GenAI labs commonly expose OPENAI_BASE_URL and OPENAI_API_BASE.
# If they are not set but the key clearly looks like a Vocareum key, use the
# standard OpenAI-compatible Vocareum gateway used in prior Udacity projects.
OPENAI_BASE_URL = os.getenv("OPENAI_BASE_URL") or os.getenv("OPENAI_API_BASE")
if OPENAI_API_KEY and OPENAI_API_KEY.startswith("voc-") and not OPENAI_BASE_URL:
    OPENAI_BASE_URL = "https://openai.vocareum.com/v1"
    os.environ["OPENAI_BASE_URL"] = OPENAI_BASE_URL
    os.environ["OPENAI_API_BASE"] = OPENAI_BASE_URL

# Chroma's OpenAI embedding function can read the key from an environment variable.
# Keep a dedicated env var too, so the collection can be reopened consistently.
CHROMA_OPENAI_API_KEY = os.getenv("CHROMA_OPENAI_API_KEY") or OPENAI_API_KEY
CHROMA_OPENAI_API_BASE = os.getenv("CHROMA_OPENAI_API_BASE") or OPENAI_BASE_URL

assert OPENAI_API_KEY, (
    "OPENAI_API_KEY was not found. Put it in .env or config.env in the notebook "
    "folder or project root, then restart/rerun the notebook."
)
assert CHROMA_OPENAI_API_KEY, "CHROMA_OPENAI_API_KEY or OPENAI_API_KEY must be set"

os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
os.environ["CHROMA_OPENAI_API_KEY"] = CHROMA_OPENAI_API_KEY

print("OPENAI_API_KEY loaded:", bool(OPENAI_API_KEY))
print("OPENAI_BASE_URL:", OPENAI_BASE_URL or "OpenAI default endpoint")
```

## Vector DB instance

```python
# PersistentClient writes the vector database to disk, so Part 2 can reuse it.
CHROMA_PATH = "chromadb"
COLLECTION_NAME = "udaplay"

chroma_client = chromadb.PersistentClient(path=CHROMA_PATH)
print(f"ChromaDB client ready at: {CHROMA_PATH}")
```

## Collection and embedding function

```python
# Keep this model consistent between indexing and querying.
embedding_kwargs = {
    "api_key_env_var": "CHROMA_OPENAI_API_KEY",
    "model_name": "text-embedding-3-small",
}

# Needed when the workspace uses an OpenAI-compatible proxy such as Vocareum.
# For a normal OpenAI account, CHROMA_OPENAI_API_BASE will be None and Chroma
# will use OpenAI's default API endpoint.
if CHROMA_OPENAI_API_BASE:
    embedding_kwargs["api_base"] = CHROMA_OPENAI_API_BASE

embedding_fn = embedding_functions.OpenAIEmbeddingFunction(**embedding_kwargs)
```

```python
# get_or_create_collection makes the notebook safe to re-run.
collection = chroma_client.get_or_create_collection(
    name=COLLECTION_NAME,
    embedding_function=embedding_fn,
    metadata={"hnsw:space": "cosine"},
)

print(f"Collection ready: {COLLECTION_NAME}")
```

## Load and format local game data

```python
def game_to_document(game: Dict) -> str:
    """Create a searchable text representation for a game JSON object."""
    return (
        f"Title: {game['Name']}\n"
        f"Platform: {game['Platform']}\n"
        f"Genre: {game['Genre']}\n"
        f"Publisher: {game['Publisher']}\n"
        f"Year of release: {game['YearOfRelease']}\n"
        f"Description: {game['Description']}"
    )


def load_game_json_files(data_dir: str = "games") -> tuple[List[str], List[str], List[Dict]]:
    """Load all local game JSON files and convert them to ChromaDB inputs."""
    data_path = Path(data_dir)
    if not data_path.exists():
        raise FileNotFoundError(f"Could not find data directory: {data_path.resolve()}")

    ids: List[str] = []
    documents: List[str] = []
    metadatas: List[Dict] = []

    for file_path in sorted(data_path.glob("*.json")):
        with file_path.open("r", encoding="utf-8") as f:
            game = json.load(f)

        doc_id = file_path.stem
        ids.append(doc_id)
        documents.append(game_to_document(game))
        metadatas.append({
            "Name": game["Name"],
            "Platform": game["Platform"],
            "Genre": game["Genre"],
            "Publisher": game["Publisher"],
            "YearOfRelease": int(game["YearOfRelease"]),
            "Description": game["Description"],
            "source_file": str(file_path),
            "source_type": "local_game_json",
        })

    return ids, documents, metadatas


ids, documents, metadatas = load_game_json_files("games")
print(f"Loaded {len(ids)} game JSON files")
print(documents[0])
```

## Add documents to ChromaDB

```python
# upsert avoids duplicate-ID errors when the notebook is re-run.
collection.upsert(
    ids=ids,
    documents=documents,
    metadatas=metadatas,
)

print(f"Collection now contains {collection.count()} documents")
```

## Semantic search check

```python
def semantic_game_search(query: str, n_results: int = 3):
    """Small helper to verify that semantic search works."""
    results = collection.query(
        query_texts=[query],
        n_results=n_results,
        include=["documents", "metadatas", "distances"],
    )

    rows = []
    for doc_id, document, metadata, distance in zip(
        results["ids"][0],
        results["documents"][0],
        results["metadatas"][0],
        results["distances"][0],
    ):
        rows.append({
            "id": doc_id,
            "distance": round(float(distance), 4),
            "name": metadata["Name"],
            "platform": metadata["Platform"],
            "publisher": metadata["Publisher"],
            "year": metadata["YearOfRelease"],
            "source": metadata["source_file"],
            "document": document,
        })
    return rows


search_results = semantic_game_search("open world Rockstar game on PlayStation 2", n_results=3)
for row in search_results:
    print(json.dumps(row, indent=2, ensure_ascii=False))
```

```python
# A simple sanity check for the submission: the vector DB should return results.
assert collection.count() >= len(ids)
assert len(search_results) > 0
print("Part 1 RAG setup completed successfully.")
```
